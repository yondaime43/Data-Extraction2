```python
necessary_inns = [] # creating a list to store necessary data

with open(path + '/necessary_inn.txt') as f: 
    for line in f:
        necessary_inns.append(line.strip()) # since there might be some spaces or \n in data, we use strip() method
```

```python
inns.dtypes # we can only select data of the same data type. 

# desired data is head_inn, but it is int64 (meanwhile data in necessary_inns is str)
```

```python
inns.head_inn = inns.head_inn.astype(str) # that is why we change it to str
```

```python
selected_inns = inns.query("head_inn in @necessary_inns") # selecting desired piece of data
```
